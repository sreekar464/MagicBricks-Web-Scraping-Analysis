```python
df["price_sqft_2019"] = df["area"].map(price_sqft_2019)
df
```

```python
df["diff_of_sqft_price"]=df["price_in_rps_per_sqft"]-df["price_sqft_2019"]
```

```python
df.groupby("area")[["price_in_rps_per_sqft","price_sqft_2019"]].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='price_in_rps_per_sqft', y='price_sqft_2019', hue='area', s=100, alpha=0.6)

# Add labels
plt.xlabel('Price per sqft')
plt.ylabel('Price per sqft in 2019')

# Show the plot
plt.title('Scatter Plot: Price per sqft vs Price per sqft in 2019 (Color = Area)')
plt.legend(title='Area')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['area'], df['price_in_rps_per_sqft'], marker='o', label='Price per sqft')
plt.plot(df['area'], df['price_sqft_2019'], marker='o', label='Price per sqft in 2019')

# Add labels and title
plt.xlabel('Area')
plt.ylabel('Price per sqft')
plt.title('Line Chart: Price per sqft vs Price per sqft in 2019 by Area')

# Add a legend
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
grouped_df = df.groupby('area')["price_in_rps_per_sqft"].mean().reset_index()
grouped_df1 = df.groupby('area')["price_sqft_2019"].mean().reset_index()
# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(grouped_df['area'], grouped_df['price_in_rps_per_sqft'], marker='o', label='Average Price per sqft')
plt.plot(grouped_df1['area'], grouped_df1['price_sqft_2019'], marker='o', label='Average Price per sqft in 2019')

# Add labels and title
plt.xlabel('Area')
plt.ylabel('Average Price per sqft')
plt.title('Line Chart: Average Price per sqft vs Average Price per sqft in 2019 by Area')

# Add a legend
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df["future_prices"] = df["price_in_rps_per_sqft"]+df["diff_of_sqft_price"]
df.info()
```

```python
grouped_df = df.groupby('area')["price_in_rps_per_sqft"].mean().reset_index()
grouped_df1 = df.groupby('area')["future_prices"].mean().reset_index()
# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(grouped_df['area'], grouped_df['price_in_rps_per_sqft'], marker='o', label='Average Price per sqft')
plt.plot(grouped_df1['area'], grouped_df1['future_prices'], marker='o', label='Prediction of average Price per sqft in future')

# Add labels and title
plt.xlabel('Area')
plt.ylabel('Average Price per sqft')
plt.title('Line Chart: Average Price per sqft vs Prediction of average Price per sqft in future by Area')

# Add a legend
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
